```python
from openai import OpenAI

# Initialize the OpenAI client with OpenRouter as the backend.
# NOTE: Replace the API key below with YOUR OWN key obtained from OpenRouter.
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="YOUR-KEY",
) 
```

```python
# Function to perform sentiment analysis using an LLM
def sentiment_analysis(text):

    # Prepare messages for the LLM
    messages = [
        {"role": "system", "content": """
            You are a sentiment analysis assistant.
            Your task is to classify the text as Positive, Negative, or Neutral.
            Respond with only one word: Positive, Negative, or Neutral.
            If unsure, say 'Manually Check'.
        """},

        {"role": "user", "content": f"Analyze the sentiment of the following text: {text}"}
    ]

    # Call the model through OpenRouter (model = "gpt-oss-20b")
    response = client.chat.completions.create(
        model="gpt-oss-20b",
        messages=messages,
    )

    # Extract and clean the model's output
    sentiment = response.choices[0].message.content.strip().capitalize()

    return sentiment

```

```python
# Sample input
input_text = "I absolutely love the new update!"

# Call the function
response = sentiment_analysis(input_text)

# Print the result
print(input_text, ': Sentiment is', response)
```
