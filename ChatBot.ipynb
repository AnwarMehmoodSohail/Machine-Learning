```python
from openai import OpenAI

# Initialize the OpenAI client with OpenRouter as the backend.
# NOTE: Replace the API key below with YOUR OWN key obtained from OpenRouter.
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="API-KEY",
) 
```

```python
# Function to create a chatbot using an LLM
def chatbot():
    print("Chatbot (type 'exit' to quit):")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "quit"]:
            print("Chatbot: Goodbye!")
            break

        # Prepare messages for the LLM
        messages = [
            {"role": "system", "content": """
                You are a friendly chatbot assistant.
                Your task is to have a natural, helpful, and polite conversation with the user.
                Respond concisely and clearly.
            """},
            {"role": "user", "content": user_input}
        ]

        # Call the model through OpenRouter (model = "gpt-oss-20b")
        response = client.chat.completions.create(
            model="gpt-oss-20b",
            messages=messages,
        )
        print("User:", user_input)
        # Extract and clean the model's output
        bot_reply = response.choices[0].message.content.strip()
        print("Chatbot:", bot_reply)

```

```python
# Start the chatbot
chatbot()
```
